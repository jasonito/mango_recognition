```python
#以python讀取及修改excel
import openpyxl

#讀取excel
wb = openpyxl.load_workbook('C:/Users/tsaiy/MyPython/Econometric/AirPollution/中山1~3月.xlsx')

#修改工作表名稱
for i in range(len(wb.sheetnames)) :
    wb.active = i
    ws = wb.active
    #月份為wb.active.title[4:6]，測項為wb.active.title[10:]
    ws.title = wb.active.title[10:]+"_"+wb.active.title[4:6]
    #print(ws.title)
    
    ws.delete_rows(34,3)
    ws.insert_cols(2)
    
    #加入測項(例如：SO2、CO、O3、PM10等等)
    #ws.max_row為36列，但有4列是不要的
    for j in range(ws.max_row-4):   
        #注意j從0開始，要加2，才會到B2儲存格
        ws['B'+str(j+2)].value = ws.title[:-3]

    #處理月份日數不齊的問題
    for j in range(1,ws.max_row+1):
        if ws.cell(j,1).value is None:
            ws.delete_rows(j,1) 
        else:
            ws['B'+str(j)].value = ws.title[:-3]

wb.save('C:/Users/tsaiy/MyPython/Econometric/AirPollution/sample.xlsx')
```
